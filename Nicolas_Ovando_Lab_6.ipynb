```python
X = cereal_data[["protein"]]
Y = cereal_data[["weight"]]
```

```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()  # code to graph a linear regression 
```

```python
linear_regressor.fit(X, Y)
Y_pred = linear_regressor.predict(X)
```

```python
plt.plot(X, Y,'.')
plt.plot(X, Y_pred, color='red')
plt.show() # this is the protein and weight graph with a red line representing the linear regression
```

```python
x = np.linspace(0,10,50) # evenly spacing the numbers 
```

```python
m = 3
b = 7 

y = m * x + b # slope formula 
```

```python
noise = 10 * np.random.randn(50,)
```

```python
y = m * x + b + noise
```

```python
plt.plot(x,y,'.') # graph for our slope + noise data 
```

```python
m_guess = 1
b_guess = 1

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.'); # here the system is taking a guess of the line of best fit
```

```python
m_guess = 5
b_guess = 15

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.'); # again the system is taking its best guess of the line but with  different m and b values
```

```python
loss = np.sum((y_guess - y)**2) # code for the losses during the systems run to find the line of best fit 
```

```python
for m_guess in np.linspace(-10,10,10):
    print(m_guess)
```

```python
for m_guess in np.linspace(-10,10,10):
    for b_guess in np.linspace(-10,10,10):
        
        y_guess = m_guess * x + b_guess

        plt.plot(x,y_guess,'r.');
        plt.plot(x,y,'g.');
        plt.show()

        loss = np.sum((y_guess - y)**2)

        print(loss,m_guess,b_guess) # the system is printing several graphs with the same data but different lines of best fit until it finds the best one
```

```python
best_loss = 100000
best_m = 0
best_b = 0 

for m_guess in np.linspace(-10,10,20):
    for b_guess in np.linspace(-10,10,20):
        
        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:
            
            best_loss = loss
            best_m = m_guess
            best_b = b_guess 

            plt.plot(x,y_guess,'r.');
            plt.plot(x,y,'g.');
            plt.show()

            print(loss,m_guess,b_guess) # again, the system is trying to find what the line of best fit will look like for the given data but this time with a bigger bet loss
```
